```python
# Iris-Setosa, Iris-Versicolour, Iris-Virginica 3가지의 붓꽃의 종류가 있다.
# 각 라벨에 대해 50행의 데이터가 있으며 전부 150개의 데이터가 있다.
# sepal length(꽃 받침의 길이):   4.3  7.9   5.84   0.83    0.7826
# sepal width(꽃 받침의 폭):    2.0  4.4   3.05   0.43   -0.4194
# petal length(꽃잎의 길이):   1.0  6.9   3.76   1.76    0.9490  (high!)
# petal width(꽃잎의 폭):    0.1  2.5   1.20   0.76    0.9565  (high!)
# 4가지 특징(특성)
# 이데이터를 이진 분류 문제로 다루기 위해서 꽃받침의 길이와 꽃받침의 폭 2가지의 특성으로 한정한다.
# 라벨은 2개로 제한한다.

# 데이터 시각화 하기
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# 데이터 플롯도를 작성
iris = datasets.load_iris()
# 꽃에 대한 특성 4개가 들어 있는 
# nbarray의 2차원 배열
X = iris.data[:,:2]
# iris.target은 0(=setosa), 1(=versicolor), 2(=virginica)이 저장되어 있는
# nbarray의 일차원 배열
y = np.where(iris.target!=0, 1, iris.target)
# np.where(조건, 참일 때 값, 거짓일 때 값)
# 결과 : 0은 0으로 1과 2는 1로 변환

plt.figure(figsize=(10,6))
plt.scatter(X[y == 0][:,0], X[y == 0][:,1], color='navy', label='0')
plt.scatter(X[y == 1][:,0], X[y == 1][:,1], color='brown', label='1')
plt.legend();
```

```python
# 시그모이드 함수와 교차 엔트로피 오차의 함수 구현하기
import numpy as np
from sklearn.datasets import load_iris

def add_intercept(X):
    # 다중 선형 회귀를 계산하기 위한 행렬을 작성
    intercept = np.ones((X.shape[0], 1))
    return np.concatenate((intercept, X), axis=1)

def sigmoid(z):
    # 시그모이드 함수
    return 1 / (1 + np.exp(-z))

def cross_entropy(h, y):
    # 교차 엔트로피 오차
    return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()

def predict_prob(X, theta):
    # θ (theta) 는 보통 모델의 파라미터(가중치) 를 의미할 때 쓰입니다.
    # 머신러닝·통계에서 아주 자주 나오는 기호
    # 교차 엔트로피 오차가 가장 작은 최신의 가중치를 이용하여 X의 데이터로부터 확률을 예측값으로 출력
    X = add_intercept(X)
    return sigmoid(np.dot(X, theta))

def predict(X, theta):
    # 시그모이드 함수에 의해 출력된 확율의 예측값은 0.5 이상은 1
    # 0.5이하는 0으로 분류한 라벨을 출력
    return predict_prob(X, theta).round()
# 이러한 함수와 iris 데이터(X,y)를 이용하여 로지스틱 회귀로 가중치를 학습하는 코드를 기술한다.
```

```python
# 로지스틱 회귀로 가중치 학습하기
# 오차를 보면서 lr을 조정한다.
# 학습률(learnig rate)을 조정하면서 가중치를 갱신하고,
# 교차 엔트로피 오차가 한없이 0에 가까워질 때의 학습률을 찾는다.

lr = 0.1
iter_nums = 300000

X = add_intercept(X)

# 가중치의 초기화
theta = np.zeros(X.shape[1])

# 최급강하법
for i in range(iter_nums):
    z = np.dot(X, theta)
    h = sigmoid(z)

    # 기울기(=편미분항)을 계산
    gradient = np.dot(X.T, (h -y)) / y.size

    # 가중치를 갱신
    theta = theta - lr * gradient

    # 교차 엔트로피 오차를 지정
    loss = cross_entropy(h, y)
    if(i % 10000 == 0 ):
        # 교차 엔트로피 오차를 출력
        print(f'loss: {loss} \t')


# 확율값 예측
iris = load_iris()
X = iris.data[:,:2]
predict_prob(X, theta)
# 확률값을 사용하여 분류 라벨을 예측
predict(X, theta)

# 최급강하법의 코드에서 loss(교차 엔트로피 오차)가 한없이 0에 가까운 theta(가중치)를 선택
# predict_prob와 predict에 인수로 넘겨 확률값과 분류 라벨을 예측한다.
```

```python
# 조금전 읽어 들인 데이터(X,y)와 계산한 가중치(theta)와 함수(predict_prob)를 
# 그래도 이용하여 분포도에 선을 긋는다. 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터를 플롯
plt.figure(figsize=(10,6))
plt.scatter(X[y == 0][:,0], X[y == 0][:,1], color='navy', label='0')
plt.scatter(X[y == 1][:,0], X[y == 1][:,1], color='brown', label='1')
plt.legend();

# x축의 스케일 폭을 정의
x_min, x_max = X[:,0].min(), X[:,0].max()
# y축의 스케일 폭을 정의
y_min, y_max = X[:,1].min(), X[:,1].max()

#  세로축과 가로축의 표시
xx, yy = np.meshgrid(np.linspace(x_min, x_max), np.linspace(y_min, y_max))
grid = np.c_[xx.ravel(), yy.ravel()]

# 분류한 선을 표시
probs = predict_prob(grid, theta).reshape(xx.shape)
plt.contour(xx, yy, probs, [0.5], linewidths=1, colors='black');
```
